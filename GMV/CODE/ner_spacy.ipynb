Importando spacy y openai

```python
import spacy
import openai
from google.colab import userdata
import os
from spacy_llm.util import assemble
import json
```

Poner como variable de entorno la clave de la api

```python
os.environ["OPENAI_API_KEY"] = userdata.get('apiKey')
```

Función que toma las entidades de un informe concreto y las ubica en un diccionario.

```python
def get_entities(nlp, report):
    doc = nlp(report)
    entities = {}
    for ent in doc.ents:
        if ent.label_ not in entities:
            entities[ent.label_] = []
            entities[ent.label_].append(ent.text)
        else:
            entities[ent.label_].append(ent.text)
    return entities
```

Función que crea un json a partir de un diccionario

```python
def create_json(entities, report_name):
  with open(f"{report_name}Entities.json", "w") as outfile:
    json.dump(entities, outfile)
```

Función que procesa un informe

```python
def process_report(report, report_name, nlp):
  entities = get_entities(nlp, report)
  create_json(entities, report_name)
```

\Función que procesa todos los informes en el json de entrada

```python
def process_input_json(file_name, nlp):
  dictionary = json.load(open(file_name))
  for key, value in dictionary.items():
    process_report(value, key, nlp)
```

Función que crea el nlp:

```python
def create_nlp():
  nlp = assemble("config.cfg")
  return nlp
```

Función que ejecuta en el orden apropiado las funciones anteriores

```python
def main():
  nlp = create_nlp()
  process_input_json("medical_reports.json", nlp)
```

Llamada a la función main

```python
main()
```
